```python
price = soup.find('div', class_='price')
```

```python
reg = price.find('span', class_='price-regular')
```

```python
def get_price(price_list):
    price_list = price_list.replace('\n', '').split()
    currency = ''
    currency.isnumeric
    price = 0
    for i in price_list:
        try:
            price = float(i)
        except Exception:
            currency = i
    return currency, price
```

```python
if reg:
    curreny, price = get_price(reg.text)
else:
    price_old = price.find('span', class_='price-old')
    price_new = price.find('span', class_='price-new')
    currency, price_old = get_price(price_old.text)
    currency, price_new = get_price(price_new.text)
```
