```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://mikroelectron.com/product/alcohol-gas-sensor-mq-3'
```

```python
response = requests.get(link).text
```

```python
soup = BeautifulSoup(response, parser='lxml', features='lxml')
```

```python
with open('mikroelectron_gas_sensor_html.html', 'w', encoding='utf-8') as f:
    f.write(soup.prettify())
```
