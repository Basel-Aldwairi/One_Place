```python
import requests
from bs4 import BeautifulSoup
import html
import re
```

```python
link = 'https://whiteangeljo.com/shop-categories/redragon-k689-keyboard/'
```

```python
response = requests.get(link).text
```

```python
# reponse = None
# with open('whiteangel_laptop.html', 'r', encoding='utf-8') as f:
#     response = f.read()
```

```python
# response
```

```python
soup = BeautifulSoup(response, parser='lxml', features='lxml')
```

```python
with open('whiteangel_laptop.html', 'w', encoding='utf-8') as f:
    f.write(soup.prettify())
```

```python
name_tag = soup.find('h1', class_='electron-summary-item electron-product-title')
```

```python
product_name = name_tag.text.strip() if name_tag else None
```

```python
product_summery = soup.find('div', class_='electron-product-summary-inner').prettify()
```

```python
product_number_tag = re.findall(r'(?<=woobt-products-)[0-9]+',product_summery)
```

```python
product_number = product_number_tag[0] if product_number_tag else None
```

```python
similar_items = soup.find('div', class_=f'woobt-products woobt-products-layout-default woobt-products-{product_number}')
```

```python
if similar_items:
    similar_items.decompose()
```

```python
# sale = soup.find('div', class_='electron-summary-item electron-product-labels')
```

```python
price_tags = soup.find_all('span', class_='woocommerce-Price-amount amount')

# print([(i, j) for i,j in enumerate(price_tags)])

currency = None
price = None
sale = None

if price_tags:
    price_tags = price_tags[:2]
    curreny_tag = soup.find('span', class_='woocommerce-Price-currencySymbol')

    currency = curreny_tag.text.strip()
    # print(currency)
    if currency:
        if currency == 'د.ا':
            currency = 'JOD'
    else:
        currency = None
    
    # for p in price_tags:
    #     sp = p.find('span')
    
    #     if sp:
    #         sp.decompose()

    price_text = price_tags[0].text.strip()
    price = re.findall(r'\d+,?\d+\.\d+',price_text)[0]
    
    on_sale = soup.find('div', class_='electron-summary-item electron-product-labels').text.strip()
    # on_sale = True
    if on_sale:
        sale_text = price_tags[1].text.strip()
        sale = re.findall(r'\d+,?\d+\.\d+',sale_text)[0]
    else:
        sale = None
```

```python
description = soup.find('div', class_='electron-small-title electron-meta-wrapper tagged_as')
```

```python
description_list = None
if description:
    tags = description.find('span', class_='meta-value')
    tags_text = tags.text
    text_list = tags_text.split(',')
    description_list = [t.strip() for t in text_list]
```

```python
# tags_text = tags.text
# text_list = tags_text.split(',')
# description_list = [t.strip() for t in text_list]
```

```python
brand = soup.find('a', class_='electron-brands')
```

```python
brand = brand.text.strip() if brand else None
```

```python
category_tag = soup.find('div', class_='electron-small-title electron-meta-wrapper posted_in')
```

```python
category = category_tag.find('span', class_='meta-value').text
```

```python
category_list_unstripped = category.split(',')
```

```python
category_list = [t.strip() for t in category_list_unstripped]
```

```python
print(f"""
{product_name=}
{product_number=}
{currency=}
{price=}
{sale=}
{description_list=}
{brand=}
{category_list=}

""")
```
