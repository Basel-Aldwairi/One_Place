```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
link = 'https://citycenter.jo/computer-hardware/components-memory-ram/adata-xpg-spectrix-d35g-8gb-rgb-ddr4-desktop-memory-3600-mhz-clock-speed-cl18-latency-supports-intel-xmp-2-0-1-35v-rohs-compliant-white'
```

```python
response = requests.get(link)
```

```python
soup = BeautifulSoup(response.text, parser='lxml',features='lxml')
```

```python
all_as = soup.find_all('a', href=True)
a = soup.find('a', href=True)
```

```python
patterns = [
    '/pc-and-laptops',
    '/computer-hardware-monitor-and-display',
    '/computer-hardware-peripherals',
    '/computer-hardware-audio',
    '/computer-hardware-storage',
    '/computer-hardware',
    '/gaming',
    '/peripherals-printer-and-scanner',
    '/electronics',
    '/Networking'
]
```

```python
visited = set()
queue = []
```

```python
def pattern_exists(link,pattern):
    return pattern in link
```

```python
for a in all_as:
    link = a.get_attribute_list('href')[0]

    valid_link = 'citycenter.jo' in link
    valid_sub = any([pattern_exists(link,pattern) for pattern in patterns])
    if valid_link:
        if valid_sub and link not in visited:
            visited.add(link)
            queue.append(link)
```
