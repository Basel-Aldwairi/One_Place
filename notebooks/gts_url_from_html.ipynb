```python
from bs4 import BeautifulSoup
import pandas as pd
import time
from multiprocessing import Pool, cpu_count
```

```python
cpu_count()
```

```python
csv_file = '../src/gts/gts_crawl.csv'
```

```python
read_time = time.time()
df = pd.read_csv(csv_file, index_col=0)
df.rename(columns= {'0' : 'html'}, inplace=True)
html_list = df['html'].to_list()
read_time = time.time() - read_time
print(f'[INFO] : CSV File Read Time = {read_time}')
```

```python
unique_urls = set()
all_urls = []
```

```python
unique_urls = set()
all_urls = []
processing_time = time.time()

for html in html_list[:]:

    soup = BeautifulSoup(html, parser='lxml', features='lxml')

    a_href = soup.find_all('a', href=True)

    for a in a_href:
        url = a.get_attribute_list('href')[0].strip()
        if 'https://gts.jo' in url:
            unique_urls.add(url)
            all_urls.append(url)

processing_time = time.time() - processing_time
print(f'[INFO] : BeautifulSoup Processing Time = {processing_time}')
```
