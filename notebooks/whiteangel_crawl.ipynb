```python
import requests
from bs4 import BeautifulSoup
import cloudscraper
import time
```

```python
current_page = "1"
max_page = 27
```

```python
data = {
    "cache":"false",
    "action" :	"electron_shop_load_more",
    "beforeSend":	"",
    "ajaxurl" :	"https://whiteangeljo.com/wp-admin/admin-ajax.php",
    "current_page":	current_page,
    "per_page" : "30",
    "max_page":	f"{max_page}",
    "cat_id" :	"",
    "brand_id" : "",
    "filter_cat":  "",
    "on_sale":"",
    "orderby":	"meta_value",
    "meta_key":	"_stock_status",
    "min_price":	"",
    "max_price":	"",
    "pstyle	":"1",
    "column":	"",
    "is_search":	"yes",
    "is_shop":	"yes",
    "is_brand":	"",
    "is_cat":	"",
    "is_tag":	"",
    "s":	"+"
}
```

```python
headers = {

    "User-Agent" : "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:144.0) Gecko/20100101 Firefox/144.0",
    "X-Requested-With ":"XMLHttpRequest",
    "Referer" : "https://whiteangeljo.com/"
    
}
```

```python
url = "https://whiteangeljo.com/wp-admin/admin-ajax.php"
response = requests.post(url, headers= headers, data=data)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, parser='lxml', features='lxml')
```

```python
url_tags = soup.find_all('a',href=True)
```

```python
urls = [url.get_attribute_list('href') for url in url_tags]
```

```python
urls_set = set()
for url_l in urls:
    url = url_l[0]

    if 'https://whiteangeljo.com' in url:
        urls_set.add(url)
```

```python
len(urls_set)
```

```python
# urls_set
```

```python
all_urls = set()
all_urls_list = []


scraper = cloudscraper.create_scraper()

for i in range(max_page + 1):
    current_page = f"{i}"

    if i % 7 == 0:
        time.sleep(0.1)

    data["current_page"] = current_page
    url = "https://whiteangeljo.com/wp-admin/admin-ajax.php"
    response = scraper.post(url, headers= headers, data=data)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, parser='lxml', features='lxml')

        url_tags = soup.find_all('a',href=True)
        urls = [url.get_attribute_list('href') for url in url_tags]
        
        for url_l in urls:
            url = url_l[0]
            if 'https://whiteangeljo.com' in url:
                all_urls.add(url)
                all_urls_list.append(url)

        print(f'[INFO {i} / {max_page}] : All URLS = {len(all_urls_list)}, Unique URLS = {len(all_urls)}')
```

```python
counting_dict = dict()
```

```python
for i in all_urls_list:
    if i not in counting_dict:
        counting_dict[i] = 1
    else:
        counting_dict[i] += 1
```
