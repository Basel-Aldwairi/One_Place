```python
import requests
from bs4 import BeautifulSoup
import re
import time
import tqdm
```

```python
base_gaming_components_url = 'https://compujordan.com/gaming-rampage/gaming-components'
```

```python
response = requests.get(base_gaming_components_url).text
```

```python
soup = BeautifulSoup(response, parser='lxml',features='lxml')
```

```python
response
```

```python
all_caption_tags = soup.find_all('div', class_='caption')
```

```python
cap = all_caption_tags[0]
```

```python
urls = []
for caption in all_caption_tags:
    a_tag = caption.find('a', href=True)
    url = a_tag.get_attribute_list('href')[0]
    urls.append(url)
```

```python
num_pages_tag = soup.find('div', class_='results')
```

```python
num_pages = int(re.findall(r'(?<=\()\d+',num_pages_tag.text)[0])
```

```python
page_url_addition = '?page='
```

```python
gaming_components_url = base_gaming_components_url + page_url_addition
```

```python
session = requests.Session()

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Referer': 'https://www.google.com',
    'Connection': 'keep-alive'
}

session.headers.update(headers)
```

```python
shop_url = 'https://compujordan.com/'
```

```python
start_time = time.time()
all_urls = {}
for i in tqdm.trange(1, num_pages + 1):
    search_url = gaming_components_url + str(i)
    response = session.get(search_url).text
    
    soup = BeautifulSoup(response, parser='lxml',features='lxml')

    breadcrumb_tag = soup.find('ul', class_='breadcrumb')
    breadcrumbs = set([li.text.strip() for li in breadcrumb_tag.find_all('li')][1:])

    all_caption_tags = soup.find_all('div', class_='caption')

    for caption in all_caption_tags:
        a_tag = caption.find('a', href=True)
        url = a_tag.get_attribute_list('href')[0]
        product_name_url = url.split('/')[-1]
        final_url = shop_url + product_name_url

        if final_url in all_urls:
            all_urls[final_url].add(breadcrumbs)
        else:
            all_urls[final_url] = breadcrumbs
            

end_time = time.time() - start_time
print(f'End time : {end_time:.2f}')
```
