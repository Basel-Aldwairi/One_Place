```python
import requests
from bs4 import BeautifulSoup
```

```python
cc_url = 'https://citycenter.jo/'
```

```python
response = requests.get(cc_url).text
```

```python
soup = BeautifulSoup(response, parser='lxml', features='lxml')
```

```python
lis = soup.find_all('li')
```

```python
li = lis[20]
```

```python
li.attr
```

```python
li.get_attribute_list('id')
```

```python
li
```

```python
valid_lis = [li for li in lis if li.get_attribute_list('id') and 'menu_url_' in li.get_attribute_list('id')[0]]
```

```python
len(valid_lis)
```

```python
urls = []

for i, li in enumerate(valid_lis):
    a = li.find_all('a')[0]
    url = a.get_attribute_list('href')[0]
    # print(url)
    if 'https://citycenter.jo' in url:
        urls.append(url)
```

```python
len(urls)
```
